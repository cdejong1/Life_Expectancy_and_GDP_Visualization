```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/cdejong1/Life_Expectancy_and_GDP_Visualization/main/Life-Expectancy-and-GDP-Starter/all_data.csv')


data = data.rename(columns = {'Life expectancy at birth (years)': 'Life_expectancy'})
print(data.head(10))
```

```python
chile_life = data.Life_expectancy[data.Country == 'Chile']
china_life = data.Life_expectancy[data.Country == 'China']
germany_life = data.Life_expectancy[data.Country == 'Germany']
mexico_life = data.Life_expectancy[data.Country == 'Mexico']
us_life = data.Life_expectancy[data.Country == 'United States of America']
zimbabwe_life = data.Life_expectancy[data.Country == 'Zimbabwe']
x_range = data.Year[data.Country == 'Chile']
country_life = [chile_life, china_life, germany_life, mexico_life, us_life, zimbabwe_life]
```

```python
plt.plot(x_range, chile_life)
plt.plot(x_range, china_life)
plt.plot(x_range, germany_life)
plt.plot(x_range, mexico_life)
plt.plot(x_range, us_life)
plt.plot(x_range, zimbabwe_life)
plt.legend(data.Country.unique())
ax = plt.subplot()
plt.title('Change in Life Expectancy by Country')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (Years)')

plt.show()
plt.clf()
```

```python
chile_gdp = data.GDP[data.Country == 'Chile']
china_gdp = data.GDP[data.Country == 'China']
germany_gdp = data.GDP[data.Country == 'Germany']
mexico_gdp = data.GDP[data.Country == 'Mexico']
us_gdp = data.GDP[data.Country == 'United States of America']
zimbabwe_gdp = data.GDP[data.Country == 'Zimbabwe']
country_gdp = [chile_gdp, china_gdp, germany_gdp, mexico_gdp, us_gdp, zimbabwe_gdp]
```

```python
for i in country_gdp:
    plt.plot(x_range, i)

plt.legend(data.Country.unique(), loc=2)
plt.title('Change in GDP by Country')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```

```python
plt.scatter(chile_life, chile_gdp)
plt.scatter(china_life, china_gdp)
plt.scatter(germany_life, germany_gdp)
plt.scatter(mexico_life, mexico_gdp)
plt.scatter(us_life, us_gdp)
plt.scatter(zimbabwe_life, zimbabwe_gdp)
plt.legend(data.Country.unique())
plt.title('Life Expectancy versus GDP by Country')
plt.xlabel('Life Expectancy at Birth (Years)')
plt.ylabel('GDP')


plt.show()
```

```python
for i in country_life:
     avg_country_life = [np.mean(i)]
     print(np.mean(i))

plt.bar('Chile',np.mean(chile_life))
plt.bar('China',np.mean(china_life))
plt.bar('Germany',np.mean(germany_life))
plt.bar('Mexico',np.mean(mexico_life))
plt.bar('USA',np.mean(us_life))
plt.bar('Zimbabwe',np.mean(zimbabwe_life))
plt.title('Average Life Expectancy from 2000 - 2016 by Country')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy at Birth (Years)')
    
plt.show()
```

After studying the data one can conclude first that the life expectancy per country has all increased each year; with the largest change occuring in zimbabwe, but still not on pace with the other countries studied.
